Буду использовать только категорию отмены

```python
df = df[df['category'] == 'CANCEL'].copy()
```

```python
df['instruction'].isna().sum()
```

```python
df['instruction'].duplicated().sum()
```

```python
df["instruction_len"] = df["instruction"].astype(str).str.len()
df["response_len"] = df["response"].astype(str).str.len()

print(df["instruction_len"].mean())
print(df["response_len"].mean())

print(df[["instruction_len", "response_len"]].describe())
```

```python
df["instruction_words"] = df["instruction"].astype(str).str.split().str.len()
df["response_words"] = df["response"].astype(str).str.split().str.len()

print(df["instruction_words"].mean())
print(df["response_words"].mean())

print(df[["instruction_words", "response_words"]].describe())
```

```python
df.sort_values(by=['instruction_words'])
```

```python
df.sort_values(by=['response_words'])
```

```python
df["instruction"] = df["instruction"].astype(str).apply(lambda x: " ".join(x.split()))
df["response"] = df["response"].astype(str).apply(lambda x: " ".join(x.split()))
```

```python
exportable_df = df.copy()
exportable_df = exportable_df[['instruction', 'response']]
```

```python
exportable_df.to_json(
    "data/dataset.jsonl",
    orient="records",
    lines=True,
    force_ascii=False,
)
```
